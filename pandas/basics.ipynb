# Learning Pandas 

Pandas is a Python Library built on top of numpy.

* The term 'Pandas' comes from 'panel data' 

* It's used to work with two major objects - Series and Dataframes

    * Series -> single, labelled column (basically a 1D array/ vector)

    * Dataframe -> multiple, labelled columns

### importing pandas

```python
import pandas as pd
```

### Creating a series

```python
data = ["100", "78", "92"]

series = pd.Series(data) #converts the data into a column, each data with ann index or 'label' and in the
print("Series : \n", series) #Prints meta data about the series in the end 
```

Customizing the index or the labels for each row/element

```python
new_data = [10, 20, 30, 40]
new_series = pd.Series(new_data, index = ['a', 'b', 'c', 'd']) # index can be any data type - tupple, list, another series, etc
print("Series with customized index : \n", new_series)
```

## Accessing and updating values 

```python
#accessing data using assigned labels
print("Data at index 'c' : ", new_series['c']) # gives error if index not found

#updating data 
new_series['c'] = 50
print("Updated series : \n", new_series)

#accessing data using numbered indices (of series without customised indices)
print("Data at index 0 : ", series[0])

#accessing data using numbered indices (of series even WITH customised indices)
print(new_series.iloc[0])
```

## Filtering

```python
data = [100, 200, 300, 400, 500, 600]
series = pd.Series(data, index = ['a', 'b', 'c', 'd', 'e', 'f'])

print("Values greater than 200 : \n", series[series > 200])
```
