# Data Cleaning

- Process of fixing and removing :
    - incomplete, incorrect or irrelevant data
- ~75% of work done with pandas is data cleaning

```python
import pandas as pd
```

### Changing data types to better suited ones

```python
df = pd.read_csv("pokemon.csv")

df["Legendary"] = df["Legendary"].astype(bool) #Changes data type to bool
print(df.to_string())
```


### Dropping/Deleting with irrelevant data

```python
## dropping columns
df = df.drop(columns=["Legendary", "No"]) #range of length of columns list : [1,n] -> 1 deletes that single column
print("Dropped Legendary and No Columns \n", df)
```

### Handling inconsistent data

```python
# remove records with missing values in the type2 col
#.dropna(subset=[list of columns]) -> drop not available based on given columns

# df = df.dropna(subset=["Type2"])
# print("Dropped empty values of type2 : \n", df) #notice the missing indices 

#replacing missing data 
#replace missing data with "None"
# #.fillna() -> fill not available
# df = df.fillna({"Type2" : "None"}) # Notice NaN replaced by None
# print(df)

# #replacing empty values with some other value
# #When we replace the NaN with "None", it is then treated as a string value and not a null one
# df = df.fillna({"Type2" : df["Type1"]}) #replaces with corresponding value in the type1 field
# print(df)

#replacing any type of values, not just the na (nit available ones)  OR fixing inconsistent values
df["Type1"] = df["Type1"].replace({"Grass" : "GRASS",
                                   "Fire" : "FIRE",
                                   "Water" : "WATER"})
print(df)
```

### Removing duplicate Entries

```python
#First adding duplicate values in the csv file and then removing them
#duplicated bulbasaur and mewtwo entries, didn't bother with updating the no column since that was already deleted

df = df.drop_duplicates()
print(df) #notice the missing indices
```
