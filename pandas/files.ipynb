# Working with csv and json files

```python
import pandas as pd
```

## Loading files

```python
#loading a .csv file

df = pd.read_csv("pokemon.csv")
print(df) #Prints the first and last five records

#print(df.to_string()) #prints as many records as it can and truncates the rest in between

#similarly, read_json("path of the file") -> loading .json file
```

## Selection Techniques

```python
#selecting a single column
print(df["Name"])  #returns a series

#accessing multiple columns
print("\n Accessing multilpe columns : \n",df[["Name", "Type1", "Weight"]])
```

```python
#Accessing single row
#method 1 : use the .loc[index] property

print(df.loc[2]) #Accessing the record at index 2

#Not very useful, but if we use the name column as the index to fetch records that would be useful
#while importing set index_col = name and now the name column is the index
new_df = pd.read_csv("pokemon.csv", index_col = "Name")
print("\n",new_df)
```

```python
#accessing row using the .loc[index] property only (or .iloc[numeric_index])
print(new_df.loc["Pikachu"])

#fetching specific columns of the given record
print("\n",new_df.loc["Charizard", ["Height", "Weight", "Type1"]])

#Slicing rows
print("\n", new_df.loc["Charizard":"Butterfree":2, ["Type1","Type2"]]) #with step 
```

### Using an input to fetch data

```python
pokemon = input("Enter a pokemon : ")
try : 
    print(new_df.loc[pokemon])
except KeyError :
    print(f"{pokemon} not found")
```

## Filtering 

Applying condition on a single column

```python
#Applying condition on a single column
tall_pokemons = df[df["Height"] > 2]
print("\n Tall pokemons : \n", tall_pokemons)

#Heavy pokemons
heavy_poks = df[df["Weight"] > 70]
print("\n Heavy Pokemons : \n", heavy_poks)
```

Applying conditions to multiple columns

```python
#Fetching water type pokemons
#either Type1 OR Type2 == water
#Each condition specifies within () for multiple conditions [with logical operators]
water_poks = new_df[(new_df["Type1"] == "Water")|
                    (new_df["Type2"] == "Water")
                    ]
print("Water Pokemons : \n", water_poks)
```

```python
#Fetching fire AND flying pokemons
ff_poks = new_df[((new_df["Type1"]  == "Fire") | (new_df["Type2"]  == "Fire")) &
                 ((new_df["Type1"]  == "Flying")| (new_df["Type2"]  == "Flying"))]
print("\n Fire and Flying pokemons : \n", ff_poks)
```
