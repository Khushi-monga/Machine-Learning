# Aggregation

```python
import pandas as pd
```

## Aggregate fuunctions
 - Reduce a set of values to a single, summary value
 - Used to summarize and analyze data
 - Often used with groupby() function


Aggregate functions with whole dataframes

```python
df = pd.read_csv("pokemon.csv")

#WORKING WITH WHOLE DATAFRAMES
# # print("Column-wise Mean : ", df.mean()) -> throws error because can't be applied to columns that don't have a numeric data type
print("Column-wise Means : \n", df.mean(numeric_only=True))
print("\n Column-wise Max : \n", df.max(numeric_only=True))
print("\n Column-wise Sum : \n", df.sum(numeric_only=True))
print("\n No of elements in each column : \n", df.count()) #since the no. of elements is to be counted, the columns don't necessarily need to be numeric type
#Notice how type2 has a count of only 67 as the null values are non counted (NaN - Not A Number)
```

With single columns

```python
print("Avg Height : ", df["Height"].mean()) #Since dealing with single column, no need to specify numeric_only
print(" Max weight : ", df["Weight"].max())
print("Total Height : ", df["Height"].sum())
print("Total Pokemons : ", df["Name"].count())
```

With multiple columns

```python
print("Avg Height  and weight: \n ", df[["Height", "Weight"]].mean())
```

### groupby()
Forms groups based on same features in a column

```python
groups = df.groupby("Type1")
print(groups) #returns an object

#applying aggregate functions groupwise
print("Avg height of all types of pokemons : \n", groups["Height"].mean())
print("No. of each type of Pokemon : \n", groups.count())
print("Sum of numeric characteristics of each type of pokemon : \n", groups.sum(numeric_only=True)) #comment the above two commands to view this one
```
