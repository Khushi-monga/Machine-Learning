# Dataframes in Pandas

A dataframe is a tabular data structure with both rows and columns (2D)

```python
import pandas as pd
```

## Creating a dataframe 

```python
#Created using a dictionary
#The keys are the column names 
#Values are lists and contain the data for that column

data = {
    "Name" : ["Sooraj", "Shikha", "Aman", "Aditi", "Kailash"],
    "Age" : [23, 27, 32, 21, 48]
}

df = pd.DataFrame(data) #df - short for data frame
print(df)

#Customising labels
df = pd.DataFrame(data, index = ["Emp1", "Emp2", "Emp3", "Emp4", "Emp5"])
print("\n Updated index dataframe : \n",df)
```

## Adding new columns and rows 

```python
#Adding a new column
df["Position"] = ["Software Engineer", "Product manager", "AI Engineer", "Data Scientist", "Business Analyst"]
print("\n Added new column \n", df)

#Adding a new row
new_data = {"Name":"Saloni", "Age":27, "Position": "Jr. Developer"}
new_row = pd.DataFrame(new_data, index =["Emp6"])
df = pd.concat([df, new_row]) #FOCUS ON THE SYNTAX
print("\n Added new row : \n", df)
```

```python
#Adding multiple rows
rows_data = [
    {"Name":"Jaishankar", "Age":38, "Position": "Marketing strategist"},
    {"Name":"Alisha", "Age":26, "Position": "Graphic Designer"}
]
new_rows = pd.DataFrame(rows_data, index = ["Emp7", "Emp8"])
df = pd.concat([df, new_rows])
print("\n Adding multiple rows : \n", df)
```
