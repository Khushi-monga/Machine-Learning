# Advance Operations with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

Sample data structure : [restaurant_id, 2021, 2022, 2023, 2024] {sales in these years respectively}

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000] #Chai Point
])

shape = sales_data.shape
print("Sales data shape: ", shape)

#sales data for first three restaurants 
print("First three restaurants : \n", sales_data[:3])

# data without the restaurant ids
print("All data : \n", sales_data[:, 1:])
```

## Summing data in the matrix, row-wise and column-wise, and other operations

```python
#calculating total yearly sales of all the restaurants combined 
#basically separate sum of individual columns

yearly_sales = np.sum(sales_data[:, 1: ], axis = 0) #axis = 0 -> columns (1-> row)
print("Yearly sales : ", yearly_sales)

#Minimum sales per restaurant 
min_sales = np.min(sales_data[:, 1:], axis = 1)
print("Min sales per restaurant : ", min_sales)

#Average sales per year
avg_sales = np.average(sales_data[:, 1:], axis = 0)
print("Average yearly sales : ", avg_sales)

#Cumulative sum
cumulative_sum = np.cumsum(sales_data[:,1:], axis = 1)
print("Cumulative sum of sales per restaurant : ", cumulative_sum)

#Monthly average calculation
monthly_average = sales_data[:, 1:]/12
print("Monthly average per restaurant : ", monthly_average)
```

## Plotting the data

```python
#Plotting all the data 
plt.figure(figsize=(10,6)) #create a figure and specify its size and other requirements
plt.plot(sales_data[:, 1:]) #Plot the specified data
plt.title("Yearly sales of five restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector operations

```python
vector1 = np.arange(1,6)
vector2 = np.arange(6,11)

#Vector Addition
vector_sum = vector1 + vector2
print("Vector sum : ", vector_sum)

#Element-wise vector multiplication
vector_prod = vector1 * vector2
print("Vector element product : ", vector_prod)

#Dit product of vectors
dot_prod = np.dot(vector1, vector2)
print("Dot product of vectors : ", dot_prod)

#Angle between the two vectors 
#angle = theta = cosInverse(a.b / |a||b|)
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between the two vectors : ", angle)
```

## Creating vectorized methods 

```python
#vectorized methods - apply an operation to each element individually
#Not needed for scalar operations, they are automatically broadcast to each element(like a*array -> each element multiplied like usually happens with matrices)
restaurant_types = np.array(['Biryani', 'Chinese', 'Pizza', 'Burger', 'Cafe'])
vectorized_upper = np.vectorize(str.upper) #the upper method of string can now be easily applied to each member of a vector easily using this new method
print("Vectorized Upper applied : ", vectorized_upper(restaurant_types))
```
