# Handling Missing Data

In a dataset there is usually the problem of missing data. There are three maajor ways of dealing with these missing values: 
1. Remove the record/ Complete Case Analysis(CCA) 
    - {though not usually preferred because it gets rid of the values present in the other cols for the same record}
2. Univariate handling (one col at a time) 

[Handled with SimpleImputer()] Consists of various techniques, classfied based on the type of missing values : 
- Numerical (impute with :)
    - mean/median val
    - random val
    - end of distribution val
- categorical (impute with :) 
    - mode
    - 'missing'
3. Multivariate handling(multiple cols simultaneously) - consists of following techniques : 
    - KNNImputer()
    - IterativeImputer()/ MICE algo

## Complete Case Analysis (CCA)
- It literally means alaysis data only for those observations for which there is information in each of the variables of the dataset
- Also called 'list-wise deletion' : consists of discarding all those observations for which values in any of the variables is missing.

### When to use CCA?
- **When the data is MCAR[Missing Completely At Random]** : if the missing data is completely random in the dataset and not clubbed together anywhere. 
- **missing data < 5% of total data** 

### Advantages of CCA
1. Easy to implement
2. Varaible distribution remains the same (if the data is MACR, then the distribution is not affected by the scatter of removal of the observations)

### Disadvantages of CCA
1. It can exclude a large fraction of the dataset if the missing data is abundant
2. The values of other values discarded could be useful
3. Model will not know how to handle missing data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/data_science_job.csv')
df.sample(5)
```

```python
# percentage of missing values in each of the cols
df.isnull().mean()*100
```

Applying CCA only to those cols which have missing data <5%

```python
# List of cols to apply cca on
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].head()
```

```python
# Check how much data remains after removing all the missing vals in the selected cols
len(df[cols].dropna())/ len(df)
```
