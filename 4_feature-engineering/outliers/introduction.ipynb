# Outliers

## What are outliers?
- Outliers are those data points in a dataset which deviate significantly from the rest of the data. They could occur due to errors or can be natural occurence as well. 
- They may impact the mean and metrics of the data and hence affect the performance of the model. 

## Effect of outliers on ML Algorithms
- Algorithms affected significantly by outliers : 
    - Linear Regression
    - Logistic Regression
    - Adaboost Ensemble algorithm
    - Deep learning

- All these are weight based algos

## How to treat outliers?
Common techniques to deal with outliers : 
1. Trimming - remove the outliers 
    - May result in thinning of data in case of large num of outliers
    - But is a fast method of dealing with outliers
2. Capping - set a range for the data and any value beyond that range is deemed an outlier and thus is replaced by the end value that it is close to

Less common methods : 
- Missing value - treat outliers as missing data and replace them as nan
- Discretization - Reduces the impact of outliers on the model by turning numeircal data into categorical data

## How to detect outliers?
Different methods for different distributions: 
1. Normal Distribution
    - Outliers if value < mean - 3(std dev) or > mean + 3(std dev)

2. Skewed distribution
    - Boxplot :
        - val < Q1 - 1.5IQR or
        - val > Q3 + 1.5IQR

3. Other Distributions
    - val < 2.5th %ile or 
    - val > 97.5th %ile
