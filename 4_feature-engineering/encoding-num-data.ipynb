# Encoding Numerical data

## Why encode numerical data ?
- Numeric Values are encoded in cases when the range of numeric data is more important than the exact numeric value in the predictions, ex age, bp reading (normal, low, high), etc. 
- It also helps in handling outliers.
- Uneven distributions can be transformed into more uniform ones 
- In some cases also helps simplify the complex patterns, making them easier for the model to learn.

## Steps to encode numeric data 
1. Convert numeric data into categorical data 
2. Perform encoding

## 1. Converting numeric data to categorical data 
There are two methods to perform this : 
1. Discretization/Binning
2. Binarization

### Discretization
Discretization is the process of converting continuous data into dicsrete values by grouping data into a set of contiguous intervals,that span the range of the variable's values. This process is also known as binning, here bin is the alternative name for an interval. 

* Advantages of Binning :-
- Handles outliers (Because they get grouped into the bins and not considered individually anymore)
- Improve the value spread (In case of unevenly distributed value spreads(PDFs), helps make them more uniform and hence improve model performance)

#### Types of Discretization
There are three ways of performing discretization : 
1. Unsupervised Discretization
    - Equal-width/ Uniform binning
    - Equal-frequency/Quantile binning
    - K-means binning
2. Supervised discretization
    - Decision trees binning (covered after decision trees)
3. Custom discretization

##### 1. Equal Width Binning
- Divides the data into equal bins of equal width
- Working :
    - specify the no. of bins to create
    - bin width = (max - min)/no. of bins
    - That many bins of equal width are created. Simple
- This handles outliers 
- Does not change the value spread

##### 2. Equal-frequency/ Quantile binning
- Divides data into bins such that each bin has equal number of data values or at equal percentile distributions.
- Working : 
    - Specify the number of bins
    - Divide the total % by number of bins => ex bins = 10 then each bin has 100/10 = 10% of the data values. 
- Handles outliers 
- Value spread becomes more uniform [uniform as in uniform distribution (reactangle block)]

##### 3. K-means Binning
- Uses the K-means clustering algorithm to group data into clusters. 

### Implementing discretization
All three can be implemented using the KBinsDiscretizer class from sklearn.preprocessing.
Three arguments : 
1. num of bins
2. strategy (uniform/quantile/kmeans)
3. encoding (ordinal/ohe)

### Custom/ Domain-based discretization
When you have to create categories based on your own knowldge or in some particular manner.  For ex : age groups - [0,18] as kids
[18,60] as working, [60,80] as elderly
The you custom define your bins. May have to apply using pandas, look into it.

## Binarization
Classifying the numeric data into two categories [0/1]. For example if age < 18 => not adult (0), >18 => adult(18)
Ex use case : converting colored img to black n white - threshold at 127.5 ([0,225]). if <127.5 => intensity of pixel = 0(black) else white (1)

Implemented using Binarizer class (sklearn.preprocessing)
Two parameters : threshold and copy= True/False (if true creates a copy of the col and makes changes in that else makes changes in the same col)
