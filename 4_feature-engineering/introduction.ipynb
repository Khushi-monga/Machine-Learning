# Feature Engineering

It is the process of using domain knowledge to extract features from raw data. These features are then used to improve the performance of machine learning algorithms. 

## Steps of Feature engineering pipeline

### 1. Missing Values Imputation
Either remove the missing values or fill the missing values, because they are not accepted during training.
While filling those values, they could be replaced by the mean, median, or mode and this fixing of missing values is known as imputation. 

### 2. Handling categorical values 
Converting string values into numerical values for better handling and understanding. 
Or converting numerical values to grouped/categorical values.

### 3. Outlier detection and removal 
Outliers deviate the model from the average trend and hence must be removed before feeding the data to the model for training.

### 4. Feature Scaling 
Features must be normalized and have standard scales so that their contribution to the prediction is decided by their weights and not their scales and quantitaive values.

### 5. Feature Construction
From the given set of features, some features are complementary and can be combined into a single, more meaningful and crisp feature. Creating new features from given features like that is feature construction.

### 6. Feature Selection
Not all features are always useful. Minimizing the number of features, while retaining the crucial information is feature selection

## Feature scaling

Feature scaling is the technique of standardizing independent features, preset in the data, in a fixed range.
There are two types of feature scaling : 
- Standardization
- Normalization

### Standardization 
#### (Also known as z-score normalization)
Basically, (x - mean)/std deviation. This results in the new data having zero mean and std devitation 1

#### When to use standardization?
K-means, KNN, PCA, ANN, Gradient descent (because they make use of distances in one way or another)

### Normalization
It is a techinque of data preparation in machine learning. The goal of normalization is to change the values of numeric columns in the dataset without distorting the differences in the ranges of values or losing information. 
Basically, removing the units from numeric data (like kg/g/pounds/ etc in weight) and bringing it down to a common scale without the unit. 

#### Types of normalization : 
- Minmax scaling (most common, even synonymous with normalization)
- Mean normalization
- Max absolute scaling
- Robust Scaling

#### Minmax scaling
Transformed value = Xi = (xi - xmin)/(xmax - xmin)

Brings down the range of values to [0,1].                                       
- min 0 -> when xi = xmin => Xi = 0                                      
- and when xi = xmax  => Xi = 1 (numr and demr cancel out)

#### Mean normalization 
Xi = (xi - mean)/(xmax - xmin)

Transformed range : [-1,1]

Used for centered data, but standardization preferred for that, so not that used.

#### Max Absolute Scaling (maxAbsScaler(), sklearn)
Xi = xi/|xmax|

Used for sparse data (data with a lot of 0s)

#### Robust Scaling (RobustScaler(), sklearn)

Xi = (xi - median)/IQR (Inter Quartile Range - 75th %ile vale - 25th %ile value)

*Robust to **OUTLIERS***
