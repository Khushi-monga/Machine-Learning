# Feature Engineering

It is the process of using domain knowledge to extract features from raw data. These features are then used to improve the performance of machine learning algorithms. 

## Steps of Feature engineering pipeline

### 1. Missing Values Imputation
Either remove the missing values or fill the missing values, because they are not accepted during training.
While filling those values, they could be replaced by the mean, median, or mode and this fixing of missing values is known as imputation. 

### 2. Handling categorical values 
Converting string values into numerical values for better handling and understanding. 
Or converting numerical values to grouped/categorical values.

### 3. Outlier detection and removal 
Outliers deviate the model from the average trend and hence must be removed before feeding the data to the model for training.

### 4. Feature Scaling 
Features must be normalized and have standard scales so that their contribution to the prediction is decided by their weights and not their scales and quantitaive values.

### 5. Feature Construction
From the given set of features, some features are complementary and can be combined into a single, more meaningful and crisp feature. Creating new features from given features like that is feature construction.

### 6. Feature Selection
Not all features are always useful. Minimizing the number of features, while retaining the crucial information is feature selection
