# Handling Date and Time variables/ Columns

Date and time values usually hold a lot of hidden information which prove to be useful and important in different kinds of analysis.

Ex : if the date is 23 August, 2012. 

Then it holds information such as : 
- Date of the day
- Month
- Year (these are the obvious ones)
- Day of the week 
- Is it a weekend
- which quarter is it
- which semester is it
and many more data points according to the need of analysis

```python
import numpy as np
import pandas as pd
```

```python
#Load the data sets
date = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

### Working with dates

```python
date.sample(5)
```

```python
date.info()
```

```python
# converting date col to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Now extracting various types of attributes from that data 
#Extracting year

date['year'] = date['date'].dt.year
print(date.head())
```

```python
# Exatracting month
date['month'] = date['date'].dt.month
print(date.head())
```

```python
# extracting month  with name
date['month'] = date['date'].dt.month_name()
print(date.head())
```

```python
# extracting date
date['day'] = date['date'].dt.day
print(date.head())
```

```python
# Extracting the day of the week
date['day_of_week'] = date['date'].dt.day_of_week
print(date.head())
```

```python
# name of the day of the week 
date['day_of_week'] = date['date'].dt.day_name() # 0 indexed the days 
print(date.head())
```

```python
# Check if it's the weekend
date['Weekend'] = np.where(date['day_of_week'].isin(['Saturday', 'Sunday']), 1, 0)
date.head()
```

```python
#extract time elapsed between two dates
import datetime 

today = datetime.datetime.today()

print((today - date['date']).head())
print("\n", (today - date['date']).dt.days.head()) # prints just the days
```

## Dealing with time

```python
# Convert to datetime datatype
time['date'] = pd.to_datetime(time['date'])
time.sample(5)
```

```python
time.info()
```

```python
# extracting time attributes - hour, mins, secs
time['hour'] = time['date'].dt.hour
time['mins'] = time['date'].dt.minute
time['secs'] = time['date'].dt.second

time.head()
```

```python
#Extract time as a whole
time['time'] = time['date'].dt.time
time.head()
```
