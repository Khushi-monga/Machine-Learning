# Frame Problem Statement

The first step in building any ML model is to clearly frame the problem and its requirements.

## Taking a real example
### Netflix Revenue
The *Business problem* is to increase revenue of Netflix
As a data scientist, you may think of three possible paths - 
- Increase the number of new customers
- Increase the subscription charges
- Reduce the number of customers leaving Netflix (also known as churn rate)

Out of the three, the third option makes the most sense and is the most implementable.
So if the current churn rate is, say, 4% then the goal may be to bring it down to 3.75 (at a large scale even this is significant).
So now we have transformed our *Business problem* into a **math problem**. This was **step 1** of framing the problem

### Step 2 : Identify the type of the problem

So now, in order to reduce the churn rate you have to first identify/predict which customers are going to leave Netflix (yes/no =
. supervised classification)
After you have identified which are the potential customers looking to leave Netflix, what are you going to do to stop them from leaving? One of the immediate solutions that comes to the mind is to offer a discount to that customer, hoping they'd stay. There could be more options, but let's proceed with this for now, for the sake of an example.

Now, even when a customer is predicted to be leaving Netflix, everyone may not have the same amount of coviction to leave netflix, some may have already decided to leave, some may be pissed off over something, some may be just thinking of leaving, etc. So why offer everyone the same amount of discount, that would be very costly. So, we also analyse the sentiment of the customers predicted to be leaving Netflix and offer a discount depending on their sentiment. For this, we can generate a score of how sure the customer is to leave, like 50% sure, 100% sure, etc. 

So now our classification problem turns into a *regression problem*. 

### Step 3 : Getting the Data
Figure what kind of data yoy require. Like in this case the required data could be : 
- Watch time
- Search but did not find
- Content left in middle
- Clicked on recommendations(order of recommendations)

Finally just check for any assumptions. Check whether the features you had thought of choosing, is that data even available or not, will the same solution or model work the same everywhere in the world?
