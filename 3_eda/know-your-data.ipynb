### 7. How is the correlation between the columns

```python
#usually used to understand teh correlation between the output col with each ip col and get rid of those cols which are not contributing significantly, or fix bias and other similar actions
# df.corr() #correlation of each col with all the others

#in our case we don't have an output field 
df2 = pd.read_csv('../ml-basics/basic-life-cycle/placement.csv')

print(df2.sample(5))

#correlation of all cols with the output col
print("\n", df2.corr()['placement'])
```
